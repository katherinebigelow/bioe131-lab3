```python
%matplotlib inline
import sys
import matplotlib.pyplot as plt
import numpy as np
```

```python
input_file = open(r"romeoandjuliet.txt")
```

```python
words = {}
for line in input_file:
    line_contents = line.split()
    for word in line_contents:
        if word in words:
            words[word]+=1
        else:
            words[word] = 1
```

```python
print(words)
```

```python
word_list = sorted(words.items(), key = lambda x: x[1])[::-1]
x, y = zip(*word_list)
```

```python
plt.figure(figsize = (20, 10))
plt.bar(x[:50], height = y[:50])
print(plt.xticks())
plt.xticks(rotation = 45)
#plt.xticks(np.arange(50), labels = x[:50], rotation = 45)
plt.xlabel("Words")
plt.ylabel("Count")
plt.title("Most common words")
```

```python
plt.figure(figsize = (20, 10))
plt.bar(x[-50:], height = y[-50:])
print(plt.xticks())
plt.xticks(rotation = 45)
#plt.xticks(np.arange(50), labels = x[:50], rotation = 45)
plt.xlabel("Words")
plt.ylabel("Count")
plt.title("Least common words")
```

## The ten most common words are "the", "I", "and", "to", "a", "of", "my", "is", "in", "that": words that are used in most sentences.
## The least common words are just the words that appear only once.
